````python
print("¡A jugar!")
pt = 0
p1= int(input("¿Cuántas veces ha ganado el Manchester United la UEFA Champions Leaugue? "))
if p1 == 3:
  print("Correcto")
  pt += 1
else:
  print("Incorrecto")

p2= int(input("¿Cuántas veces ha ganado España la Eurocopa? "))
if p2 == 4:
  print("Correcto")
  pt += 1
else:
  print("Incorrecto")

p3= int(input("""¿Qué jugador es el máximo goleador histoŕico del FC Barcelona?
1. Luis Suárez
2. Lionel Messi
3. Samuel Eto'o
4. Hristo Stoichkov
 """))
if p3 == 2:
  print("Correcto, es Messi con 672 goles")
  pt += 1
else:
  print("Incorrecto")

p4= int(input("""¿Qué equipo ha ganado más veces la UEFA Europa League?
1. Atlético de Madrid
2. Tottenham Hotspur
3. Atalanta
4. Sevilla FC
 """))
if p4 == 4:
  print("Correcto, el Sevilla la ha ganado en 7 ocasiones")
  pt += 1
else:
  print("Incorrecto")

p5= int(input("""¿Qué selección ha ganado más mundiales?
1. Brasil
2. Italia
3. Alemania
4. Argentina
 """))
if p5 == 1:
  print("Correcto, Brasil lo ha ganado en 5 ocasiones")
  pt += 1
else:
  print("Incorrecto")

if pt == 5:
  print("Has acertado 5 preguntas, tienes un 100% de aciertos")
elif pt == 4:
  print("Has acertado 4 preguntas, tienes un 80% de aciertos")
elif pt == 3:
  print("Has acertado 3 preguntas, tienes un 60% de aciertos")
elif pt == 2:
  print("Has acertado 2 preguntas, tienes un 40% de aciertos")
elif pt == 1:
  print("Has acertado 1 preguntas, tienes un 20% de aciertos")
else:
  print("Has acertado 0 preguntas, tienes un 0% de aciertos")




````
